# Feature Engineering
- data from '1993-11-12' to '2020-03-14'

```python
df = pd.read_csv(data_dir / 'preprocessed_data.csv')
df['date'] = pd.to_datetime(df['date'])
df.shape
```

```python
df.head()
```

```python
# !!! get latest data for usman
name1 = 'Dominick Reyes'
fighter1 = df[df['R_fighter'] == name1].append(df[df['B_fighter'] == name1])
fighter1 = fighter1[fighter1['date'] == fighter1['date'].max()]
fighter1
```

```python
# !!! get latest data for Jorge Masvidal
name2 = 'Jan Blachowicz'
fighter2 = df[df['R_fighter'] == name2].append(df[df['B_fighter'] == name2])
fighter2 = fighter2[fighter2['date'] == fighter2['date'].max()]
fighter2
```

```python
# now we want to consolidate them into one row, with one fighter as red, the other as blue
fighter1 = fighter1[[x for x in df.columns if x.startswith('B_')]].reset_index(drop=True)
fighter1.head(1)
```

```python
fighter1.columns = fighter1.columns.str.strip('B_')
fighter1 = fighter1.add_prefix('R_')
fighter1.columns
```

```python
fighter2 = fighter2[[x for x in df.columns if x.startswith('B_')]].reset_index(drop=True)
fighter2.columns = [x for x in df.columns if x.startswith('B_')]
fighter2
```

```python
fight1 = fighter1.merge(fighter2, left_index=True, right_index=True)
fight1
```

```python
fight1.B_fighter
```

```python
mycols = [x for x in df.columns if x.startswith('R_')] + [x for x in df.columns if x.startswith('B_')]
mycols
```

```python
# drop other columns
fight1=fight1[mycols]
fight1.drop(columns=['R_fighter','B_fighter'], inplace=True)

df=df[mycols + ['Winner']]
df.drop(columns=['R_fighter','B_fighter'], inplace=True)
```

```python
# next, we want to subtract Blue from Red, so that we capture both fighters in a single variable
def consolidate_vars(df):
    r_cols = sorted([x for x in df.columns if x.startswith('R_')])
    b_cols = sorted([x for x in df.columns if x.startswith('B_')])
    new_cols = sorted([x.strip('R_') for x in r_cols])
    for col in range(0, len(r_cols)):
        df[new_cols[col]] = df[r_cols[col]] - df[b_cols[col]]
    df = df.drop(columns= r_cols + b_cols)
    return df      
    
```

```python
# train data!
new_df = consolidate_vars(df)
```

```python
# test data! 
fight1 = consolidate_vars(fight1)
fight1.shape
```

# Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.model_selection import cross_val_score
```

```python
y_label = new_df.pop('Winner')
x_label = new_df
```

```python
# train/test split
x_train, x_test, y_train, y_test = train_test_split(x_label, y_label, test_size=0.10, random_state=42)
print(y_train.shape)
print(y_test.shape)
```

## Random Forest

```python
clf = RandomForestClassifier(n_estimators = 500, random_state=42)
model = clf.fit(x_train, y_train)
pred = model.predict(x_test)
print('Accuracy of ', accuracy_score(pred, y_test)*100)
```

```python
f1_score(y_test, pred, average='weighted')
```

```python
feature_importances = pd.DataFrame(model.feature_importances_, index = x_train.columns, columns=['importance']).sort_values('importance', ascending=False)
feature_importances
```

```python
# Random Forest composed of 500 decision trees. We optimized parameters using cross-validation and GridSearch tool paired together
classifier = RandomForestClassifier(n_estimators = 250, 
                                    criterion = 'entropy', 
                                    max_depth = 20, 
                                    min_samples_split = 2,
                                    min_samples_leaf = 3, 
                                    random_state = 0
                                   )
classifier.fit(x_train, y_train)

#We use cross-validation with 10-folds to have a more precise accuracy (reduce variation)
accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 10)
print('Accuracy mean : ', accuracies.mean())
print('Accuracy standard deviation : ', accuracies.std())

#print('Training accuracy : ', classifier.score(X_train, y_train))
y_pred = classifier.predict(x_test)
print('Testing accuracy : ', accuracy_score(y_test, y_pred), '\n')

target_names = ["Red", "Blue"]
print(classification_report(y_test, y_pred, labels = target_names))
```
